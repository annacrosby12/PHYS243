```python
samples = 80000 # Setting up Monte Carlo simulation for 80000 samples

store_samples = np.zeros(samples) # Creating "empty" storage

for i in range(samples):     # Monte Carlo simulation
    store_samples[i] = drop_simulation()


np.average(store_samples) # Finding "average" probability of falling in lines

# Value of pi: 2l/Pt

pi_value = (2*needle_length)/(3*np.average(store_samples))
print(pi_value)

# 3.1435... 
# Not bad. Tested with 1000 samples and got ~3.112.
```

```python
needleCounts = list(map(lambda x: int(x), np.linspace(100, 10**4, 3)))
```

```python
bShortNeedleSim = bn.BuffonNeedleSimulation()
```

```python
# l and t values such that l <= t (sn = short needle)
snT = [10 for i in range(10)]
snL = [i+1 for i in range(10)]
bShortNeedleSim.simulate(snT, snL, needleCounts, True)
```

## Part 3 - Long Needle

```python
bLongNeedleSim = bn.BuffonNeedleSimulation()
```

```python
# l and t values such that l > t (ln = long needle)
lnT = [2 for i in range(20)]
lnL = [i+1 for i in range(20)]
bLongNeedleSim.simulate(lnT, lnL, needleCounts, True)
```

```python
# Clear the axis so that we won't be adding the same plots again
axes[0].clear()
axes[1].clear()

# Plot analytic formulas for long needle
bLongNeedleSim.plot(isNumericalSim=True, axis=axes[0])
bLongNeedleSim.plot(isNumericalSim=False, axis=axes[1])
```

For short needle, as the (l/t) ratio increases, the probability reaches to $\frac{2l} {t\pi}$ and the graph becomes linear.
